```python
import pandas as pd
df_path = '../productstar_df.csv'
df = pd.read_csv(df_path)
df
```

```python
df['shipped_at'] = pd.to_datetime(df['shipped_at'])
df['completed_at'] = pd.to_datetime(df['completed_at'])
```

```python
df.dtypes
```

```python
df_compl = df[df['order_state'] == 'complete']
```

```python
df_compl['day'] = df['completed_at'].dt.round('D')
df_compl
```

```python
df_compl.groupby('day').sum()[['item_count']].plot()
```

```python
df_compl = df_compl[(df_compl['day']>='2019-01-01') & (df_compl['day']<'2020-01-01')]
df_compl
```

```python
df_compl.groupby('day').sum()[['item_count']].plot()
```

```python
df_abc = df_compl[['store_id','item_count']].groupby('store_id').sum().sort_values(by='item_count', ascending=False).assign(prc = lambda x: x['item_count']/x['item_count'].sum())[['prc']].cumsum()
df_abc['abc'] = 'C'
df_abc.loc[df_abc['prc']<0.75, 'abc'] = 'B'
df_abc.loc[df_abc['prc']<0.5, 'abc'] = 'A'
df_abc
```

```python
df_t = df_compl[['store_id', 'day','item_count']].groupby(['store_id', 'day'], as_index=False).sum()
df_xyz = pd.concat([df_t.groupby('store_id').mean(),df_t.groupby('store_id').std()], axis=1).assign(prc = lambda x: x.iloc[:, 1]/x.iloc[:, 0])[['prc']]
df_xyz.hist()
```

```python
df_xyz['xyz'] = 'Z'
df_xyz.loc[df_xyz['prc']<0.8, 'xyz'] = 'Y'
df_xyz.loc[df_xyz['prc']<0.55, 'xyz'] = 'X'
df_xyz
```

```python
pd.concat([df_abc, df_xyz], axis=1).sort_values(by=['abc','xyz'])
```
